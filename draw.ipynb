```python
import matplotlib.pyplot as plt
import librosa

import glob
data_path = glob.glob("/home/zyy/cv-course/data/savee/*.wav")
data_path = [_.split("/")[-1] for _ in data_path]

sr = librosa.load(f"/home/zyy/cv-course/data/savee/{data_path[0]}", sr=None)[1]

# 绘制所有的情感分布

def get_emo(path):
    return path.split("_")[-1][:-6]

d = {}
for path in data_path:
    emo = get_emo(path)
    if emo not in d:
        d[emo] = []
    d[emo].append(librosa.load(f"/home/zyy/cv-course/data/savee/{path}")[0].shape[0])
    
d = {k: sum(v)/len(v)/sr for k, v in sorted(d.items(), key=lambda item: item[0])}

# 饼图
import seaborn as sns
# 柱状图
plt.figure(figsize=(10, 10))
# 使用seaborn绘制彩图
sns.barplot(x=list(d.keys()), y=list(d.values()))
plt.show()

def get_emo(path):
    return path.split("_")[-1][:-6]

def get_actor(path):
    return path.split("_")[0]

d = {}
for path in data_path:
    emo = get_emo(path)
    actor = get_actor(path)
    if actor not in d:
        d[actor] = {}
    if emo not in d[actor]:
        d[actor][emo] = []
    d[actor][emo].append(librosa.load(f"/home/zyy/cv-course/data/savee/{path}")[0].shape[0])
    
for k,v in d.items():
    d[k] = {k: sum(v)/len(v)/sr for k, v in sorted(d[k].items(), key=lambda item: item[0])}
    
# 柱状图
plt.figure(figsize=(10, 10))
index = 0
for k,v in d.items():
    index += 1
    plt.subplot(2, 2, index)
    plt.bar(list(v.keys()), list(v.values()))
    plt.title(f"Actor {k}")
plt.legend()
plt.show()
```
